```python
import pandas as pd
import numpy as np
```

```python
meta_Data = pd.read_csv('metadata.csv')
```

```python
meta_Data
```

```python
meta_Data = meta_Data.set_index('id')
```

```python
meta_Data
```

```python
series_for = pd.Series(index = meta_Data.index)
series_against = pd.Series(index = meta_Data.index)

for iid in meta_Data.index:
    try:
        main_point = open('Main Points/main_points_'+iid).read()
        main_point_list = main_point.split('\n')
        main_point_list.remove('')
        break_point = main_point_list.index('Against The Motion')
        col_1 = main_point_list[1:break_point]
        col_2 = main_point_list[break_point+1:]

        series_for[iid] = col_1
        series_against[iid] = col_2
    except IOError as e: 
        continue
```

```python
series_for = series_for[series_for.apply(lambda x: x != [])]
series_against = series_against[series_against.apply(lambda x: x != [])]
series_for = series_for[series_for.apply(lambda x: x != ['Coming soon!'])]
series_against = series_against[series_against.apply(lambda x: x != ['Coming soon!'])]
```

```python
series_for = series_for[~series_for.isna()]
series_against = series_against[~series_against.isna()]
```

```python
meta_Data['For_Main_Points'] = series_for
meta_Data['against_Main_Points'] = series_against
```

```python
meta_Data.to_csv('metadata_appended_main_points.csv')
```
